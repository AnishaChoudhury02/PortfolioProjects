```python
#import libraries
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

#read the data
df=pd.read_csv(r'/Users/anishachoudhury/Desktop/Anisha/portfolio 1/movies.csv')
#lets look at the data
df.head()
```

```python
#find any missing data
for col in df.columns:
    missingVal= np.mean(df[col].isnull())
    print('{} - {}%'.format(col,round(missingVal*100)))
```

```python
#type of data
df.dtypes
```

```python
#change data type for budget and gross
df = df.fillna(0)
df['budget']= df['budget'].astype('int64')
df['gross']= df['gross'].astype('int64')
```

```python
#order by revenue
df=df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
#drop any duplicates
df.drop_duplicates().head()
```

```python
#display the entire table
#pd.set_option('display.max_rows',None)
df.head()
```

```python
#scatterplot with budget vs gross revenue
plt.scatter(x=df['budget'],y=df['gross'],color='purple')
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget (in millions)')
plt.ylabel('Gross (in millions)')
plt.show()
```

```python
#plot budget vs gross using seaborn
sns.regplot(x='budget',y='gross',data=df,scatter_kws={'color':'purple'},line_kws={'color':'c'})
```

```python
#correlation
df=df.select_dtypes(include=['number'])
df.corr(method='pearson')
```

```python
#Visualisation
corr_matrix= df.corr(method='pearson')
sns.heatmap(corr_matrix, annot=True )
plt.title('Correlation Matrix for Numerical Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
#dataframe visualise with original dataset
cdf=pd.read_csv(r'/Users/anishachoudhury/Desktop/Anisha/portfolio 1/movies.csv')
cdf=cdf.sort_values(by=['gross'], inplace=False, ascending=False)
cdf_numerical= cdf
for col_name in cdf_numerical.columns:
    if (cdf_numerical[col_name].dtype=='object'):
        cdf_numerical[col_name]=cdf_numerical[col_name].astype('category')
        cdf_numerical[col_name]=cdf_numerical[col_name].cat.codes
cdf_numerical.head()
```

```python
corr_matrix= cdf_numerical.corr(method='pearson')
plt.figure(figsize = (16,5))
sns.heatmap(corr_matrix, annot=True, linewidths=0.5 )
plt.title('Correlation Matrix for Numerical Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
#filter and organise
corr_mat=cdf_numerical.corr()
corr_pairs=corr_mat.unstack()
#first 15
corr_pairs.head(15)
```

```python
sorted_pairs=corr_pairs.sort_values()
#first 15
sorted_pairs.head(15)
```

```python
high_corr=sorted_pairs[(sorted_pairs>0.5)]
high_corr
```
